## Training a Neural Network

---

### Network Structure

- **N-dimensional input data**:
  - The **input layer** will consist of **N neurons**.
- **M distinct classes** in the training data:
  - The **output layer** will consist of **M neurons**.
- The layers between the input and output are called **hidden layers**.
- A **simple neural network** consists of just a few layers, while a **deep neural network** has many layers.

---

### Steps for Training

1. **Collect training data** and label it appropriately.
2. Each **neuron acts as a simple function** that helps the network learn.
3. The neural network **trains itself** by minimizing the **error**.
   - The **error** is the difference between the predicted output and the actual output.
4. Based on the error size, the network adjusts and **re-trains** until the error is below a certain threshold.

---

### Training Process

- The neural network continuously adjusts and refines its parameters, getting closer to the solution with each iteration.
- **Learning rate** and optimization techniques are used to update the weights during the training process.

## Building a Perceptron-based Classifier

---

### Perceptron

- **Basic unit of a neural network**: A single neuron.
- **Input**: Takes inputs, computes the **weighted sum**, adds a **bias**, and produces an output.
- **Decision-making**: Uses a **linear function** for decision-making.

---

### Complexity

- **Simple Perceptrons** form the foundation of more **complex deep networks**.
  - Perceptrons are the building blocks for deep neural networks, where multiple perceptrons are stacked together in layers.

---

### Tool

- **NeuroLab**: Use NeuroLab to build a perceptron-based classifier in Python.


````python
#!pip install neurolab
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
data = iris.data  # Features
labels = iris.target  # Labels (target)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)

# Define the multilayer perceptron classifier
# Here, hidden_layer_sizes=(10,) means 10 neurons in the hidden layer
clf = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, activation='relu', solver='adam', random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Plot the training progress (training error)
plt.figure()
plt.plot(clf.loss_curve_)
plt.xlabel('Number of iterations')
plt.ylabel('Training error')
plt.title('Training Error Progress')
plt.grid()
plt.show()

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Print the accuracy
print("Accuracy on test set:", accuracy_score(y_test, y_pred))

# Display some predictions
print("\nSample predictions:")
for i in range(5):  # Displaying 5 predictions
    print(f"Input features: {X_test[i]}, Predicted class: {y_pred[i]}, True class: {y_test[i]}")
````

## Constructing a Single Layer Neural Network

---

### Perceptron Limitations

- A **single perceptron** is too simple to solve complex tasks.
  - It can only solve problems that are **linearly separable**.

---

### Single Layer Neural Network

- A **single layer neural network (NN)** consists of a set of **independent neurons** working together.
- Each neuron processes the input data and contributes to the network's overall output.

---

### Goal

- The goal of a single layer neural network is to **process input data** and produce the correct **output**.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
data = iris.data[:, :2]  # Use the first two features (Dimensions 1 and 2)
labels = iris.target  # The class labels

# Plot input data
plt.figure()
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='coolwarm', edgecolors='k')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Input data')
plt.show()

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)

# Define the multilayer perceptron classifier
clf = MLPClassifier(hidden_layer_sizes=(2,), max_iter=1000, activation='relu', solver='adam', random_state=42)

# Train the neural network
clf.fit(X_train, y_train)

# Plot the training error progress
# scikit-learn doesn't have direct error progress plotting, but we can access it via loss_curve_
plt.figure()
plt.plot(clf.loss_curve_)
plt.xlabel('Number of iterations')
plt.ylabel('Training error')
plt.title('Training error progress')
plt.grid()
plt.show()

# Evaluate the model on the test set
y_pred = clf.predict(X_test)
print(f"Accuracy on test set: {accuracy_score(y_test, y_pred)}")

# Define some sample test datapoints and run the network on those points
print('\nTest results:')
data_test = [[5.1, 3.5], [6.5, 2.8], [7.1, 3.0]]
for item in data_test:
    prediction = clf.predict([item])[0]
    print(f"{item} --> Predicted class: {prediction}")
````

## Constructing a Multilayer Neural Network

---

### Need for Multiple Layers

- A **single layer neural network (NN)** has **limited accuracy** and struggles to solve complex tasks.

---

### Multilayer Neural Network

- **More layers** in the network allow it to **extract complex patterns** and learn hierarchical features.
- With multiple layers, the network can handle more abstract and intricate data representations.

---

### Goal

- The goal is to **build a multilayer network** that achieves **higher accuracy** by capturing more complex patterns in the data.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

# Load the California housing dataset
california = fetch_california_housing()
X = california.data[:, :1]  # Use only the first feature (similar to x in your original example)
y = california.target  # House prices as labels

# Normalize the y values
y /= np.linalg.norm(y)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a multilayer neural network for regression
# The first hidden layer has 10 neurons, second hidden layer has 6 neurons
# The output layer consists of 1 neuron (for regression task)
regressor = MLPRegressor(hidden_layer_sizes=(10, 6), max_iter=2000, activation='relu', solver='adam', random_state=42)

# Train the neural network
regressor.fit(X_train, y_train)

# Predict on the training data
y_pred_train = regressor.predict(X_train)

# Evaluate the model on the test set
y_pred_test = regressor.predict(X_test)

# Create the subplots: 3 plots in one row
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot the input data (first feature vs. house prices)
axes[0].scatter(X_train, y_train)
axes[0].set_xlabel('Feature 1 (e.g., MedInc)')
axes[0].set_ylabel('House Price (normalized)')
axes[0].set_title('Input Data')

# Plot the training error progress
axes[1].plot(regressor.loss_curve_)
axes[1].set_xlabel('Number of iterations')
axes[1].set_ylabel('Training error')
axes[1].set_title('Training Error Progress')
axes[1].grid()

# Plot the predicted vs actual output
axes[2].plot(X_train, y_train, '.', label='Actual')
axes[2].plot(X_train, y_pred_train, 'p', label='Predicted')
axes[2].set_xlabel('Feature 1')
axes[2].set_ylabel('Normalized House Price')
axes[2].set_title('Actual vs Predicted')
axes[2].legend()

# Adjust the layout to avoid overlap
plt.tight_layout()

# Show all the plots
plt.show()

# Print test set predictions
print(f"Test set predictions:\n{y_pred_test}")
````

## Building a Vector Quantizer

---

### Vector Quantization

- **Vector Quantization** represents input data with a fixed set of **representative points** (N-dimensional rounding).
- It maps the input data to a set of **codebook vectors** that represent the most important features.

---

### Applications

- **Image recognition**: Helps reduce the dimensionality of image data while preserving key features.
- **Semantic analysis**: Used in natural language processing (NLP) to cluster similar words or phrases.
- **Data science**: Applied in clustering and feature extraction tasks to simplify complex data.

---

### Goal

- The goal is to **build a vector quantizer** using **artificial neural networks** to represent and compress data efficiently.


````python
#!pip install numpy==1.23.0
````

````python
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl

# Load input data
text = np.loadtxt('data_vector_quantization.txt')

# Separate it into data and labels
data = text[:, 0:2]
labels = text[:, 2:]

# Define a neural network with 2 layers:
# 10 neurons in the input layer and 4 neurons in the output layer
num_input_neurons = 10
num_output_neurons = 4
weights = [1/num_output_neurons] * num_output_neurons
nn = nl.net.newlvq(nl.tool.minmax(data), num_input_neurons, weights)

# Train the neural network using the training data
_ = nn.train(data, labels, epochs=500, goal=-1)

# Create the input grid
xx, yy = np.meshgrid(np.arange(0, 10, 0.2), np.arange(0, 10, 0.2))
xx.shape = xx.size, 1
yy.shape = yy.size, 1
grid_xy = np.concatenate((xx, yy), axis=1)

# Evaluate the input grid of points
grid_eval = nn.sim(grid_xy)

# Define the 4 classes
class_1 = data[labels[:, 0] == 1]
class_2 = data[labels[:, 1] == 1]
class_3 = data[labels[:, 2] == 1]
class_4 = data[labels[:, 3] == 1]

# Define X-Y grids for all the 4 classes
grid_1 = grid_xy[grid_eval[:, 0] == 1]
grid_2 = grid_xy[grid_eval[:, 1] == 1]
grid_3 = grid_xy[grid_eval[:, 2] == 1]
grid_4 = grid_xy[grid_eval[:, 3] == 1]

# Plot the outputs
plt.plot(class_1[:, 0], class_1[:, 1], 'ko',
         class_2[:, 0], class_2[:, 1], 'ko',
         class_3[:, 0], class_3[:, 1], 'ko',
         class_4[:, 0], class_4[:, 1], 'ko')

plt.plot(grid_1[:, 0], grid_1[:, 1], 'm.',
         grid_2[:, 0], grid_2[:, 1], 'bx',
         grid_3[:, 0], grid_3[:, 1], 'c^',
         grid_4[:, 0], grid_4[:, 1], 'y+')

plt.axis([0, 10, 0, 10])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Vector Quantization')
plt.show()
````

## Analyzing Sequential Data Using Recurrent Neural Networks

---

### Sequential Data

- Unlike **static data**, **sequential data** (like **time-series**) requires special handling.
- Sequential data has **temporal dependencies** that need to be captured over time.

---

### Recurrent Neural Networks (RNNs)

- **Recurrent Neural Networks (RNNs)** are specifically designed to handle **sequential data**.
- RNNs are able to **capture temporal dependencies**, making them ideal for time-series analysis, language modeling, and more.

---

### Time-Series

- A **common type** of sequential data.
- Time-series data involves observations made at **regular time intervals** (e.g., stock prices, weather data).

---

### Goal

- The goal is to **build a model for sequential data** using **RNNs**, enabling the model to learn from patterns over time.


````python
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl

# Define a function to generate the waveforms
def get_data(num_points):
    # Create sine waveforms
    wave_1 = 0.5 * np.sin(np.arange(0, num_points))
    wave_2 = 3.6 * np.sin(np.arange(0, num_points))
    wave_3 = 1.1 * np.sin(np.arange(0, num_points))
    wave_4 = 4.7 * np.sin(np.arange(0, num_points))

    # Create varying amplitudes
    amp_1 = np.ones(num_points)
    amp_2 = 2.1 + np.zeros(num_points)
    amp_3 = 3.2 * np.ones(num_points)
    amp_4 = 0.8 + np.zeros(num_points)

    # Create the overall waveform
    wave = np.array([wave_1, wave_2, wave_3, wave_4]).reshape(num_points * 4, 1)
    amp = np.array([[amp_1, amp_2, amp_3, amp_4]]).reshape(num_points * 4, 1)

    return wave, amp

# Define a function to visualize the output of the neural network
def visualize_output(nn, num_points_test):
    wave, amp = get_data(num_points_test)
    output = nn.sim(wave)
    plt.plot(amp.reshape(num_points_test * 4))
    plt.plot(output.reshape(num_points_test * 4))

# Define the main function and create a waveform
if __name__=='__main__':
    # Create some sample data
    num_points = 40
    wave, amp = get_data(num_points)

    # Create a recurrent neural network with 2 layers
    nn = nl.net.newelm([[-2, 2]], [10, 1], [nl.trans.TanSig(), nl.trans.PureLin()])

    # Set the init functions for each layer
    nn.layers[0].initf = nl.init.InitRand([-0.1, 0.1], 'wb')
    nn.layers[1].initf = nl.init.InitRand([-0.1, 0.1], 'wb')
    nn.init()

    # Train the recurrent neural network
    error_progress = nn.train(wave, amp, epochs=1200, show=100, goal=0.01)

    # Run the training data through the network
    output = nn.sim(wave)

    # Plot the results
    plt.subplot(211)
    plt.plot(error_progress)
    plt.xlabel('Number of epochs')
    plt.ylabel('Error (MSE)')

    plt.subplot(212)
    plt.plot(amp.reshape(num_points * 4))
    plt.plot(output.reshape(num_points * 4))
    plt.legend(['Original', 'Predicted'])

    # Testing the network performance on unknown data
    plt.figure()
    plt.subplot(211)
    visualize_output(nn, 82)
    plt.xlim([0, 300])

    plt.subplot(212)
    visualize_output(nn, 49)
    plt.xlim([0, 300])

    plt.show()
````

## Building an Optical Character Recognition (OCR) Engine

---

### Goal

- **Build an OCR system** using **artificial neural networks** to recognize handwritten text.

---

### Steps

1. **Create a Python file** for the project.
2. **Import necessary packages** for building the model (e.g., TensorFlow, NumPy, etc.).


````python
import numpy as np
import neurolab as nl

# Define the input file
input_file = 'letter.data'  # Replace with the correct path to your file

# Define the number of datapoints to be loaded from the input file
num_datapoints = 50

# String containing all the distinct characters
orig_labels = 'omandig'

# Compute the number of distinct characters
num_orig_labels = len(orig_labels)

# Define the training and testing parameters
num_train = int(0.9 * num_datapoints)  # 90% for training
num_test = num_datapoints - num_train  # 10% for testing

# Define the dataset extraction parameters
start = 6
end = -1

# Creating the dataset
data = []
labels = []

with open(input_file, 'r') as f:
    for line in f.readlines():
        # Split the current line tabwise
        list_vals = line.split('\t')

        # Check if the label is in our ground truth labels. If not, we should skip it.
        if list_vals[1] not in orig_labels:
            continue

        # Extract the current label and append it to the main list
        label = np.zeros((num_orig_labels, 1))
        label[orig_labels.index(list_vals[1])] = 1
        labels.append(label)

        # Extract the character vector and append it to the main list
        cur_char = np.array([float(x) for x in list_vals[start:end]])
        data.append(cur_char)

        # Exit the loop once the required dataset has been created
        if len(data) >= num_datapoints:
            break

# Convert the data and labels to numpy arrays
data = np.asfarray(data)
labels = np.array(labels).reshape(num_datapoints, num_orig_labels)

# Extract the number of dimensions
num_dims = len(data[0])

# Create a feedforward neural network with 128 neurons in the first layer, 16 in the second, and output layer with `num_orig_labels` neurons
nn = nl.net.newff([[0, 1] for _ in range(num_dims)], [128, 16, num_orig_labels])

# Set the training algorithm to gradient descent
nn.trainf = nl.train.train_gd

# Train the neural network
error_progress = nn.train(data[:num_train, :], labels[:num_train, :], epochs=10000, show=100, goal=0.01)

# Predict the output for test inputs
print('\nTesting on unknown data:')
predicted_test = nn.sim(data[num_train:, :])

for i in range(num_test):
    print('\nOriginal:', orig_labels[np.argmax(labels[num_train + i])])
    print('Predicted:', orig_labels[np.argmax(predicted_test[i])])
````

## What are Convolutional Neural Networks (CNNs)

---

### Neural Networks (NNs)

- Composed of **neurons** with **weights** and **biases**.
- **Trained** to make predictions from input data.

---

### Deep Networks

- **Multiple layers** of neurons.
- Enables **Deep Learning** through hierarchical feature extraction.

---

### Challenge with Traditional NNs

- **Input data is flattened** (e.g., 2D images become 1D vectors).
- **Spatial information** (pixel relationships) is **lost**.
- Limits performance on structured data like images.

---

### Convolutional Neural Networks (CNNs)

- **Preserve the 2D structure** of images.
- **Special layers** (like convolution and pooling) process **spatial features**.
- **More efficient and effective** for image-related tasks.

---

### Goal of CNNs

- **Map raw images  correct class labels**.


## Architecture of CNNs

---

### Ordinary Neural Networks (NNs)

- **Inputs are flattened** into 1D vectors.
- Use **fully connected layers**:
  - Each neuron connects to **every neuron** in the previous layer.
- **Huge number of weights**:
  - Example: A **2562563** image  **196,608** inputs for a **single neuron**.
- **Not scalable** for large images:
  - High memory and computational demands.

---

### Convolutional Neural Networks (CNNs)

- **Inputs remain 3D** (width  height  depth).
- **Local connectivity**:
  - Each neuron connected only to a **small region** of the input.
- **Filters (kernels)**:
  - Small **NxN matrices** slide over the input to **extract features**.
- **Multiple filters**:
  - **M different filters** capture different features (e.g., **edges**, **textures**, **corners**).
- **Deeper layers**:
  - Capture **higher-level features** like shapes, objects, etc.



## Types of Layers in a CNN

---

### 1. Input Layer

- **Takes raw image data** as input.
- The data remains in its **original 3D shape** (width  height  depth).

---

### 2. Convolutional Layer

- **Applies filters** (kernels) to the input image.
- Performs a **dot product** between the filter and patches of the image to extract features.

---

### 3. ReLU Layer

- **Adds non-linearity** using an activation function (e.g., `max(0, x)`).
- **ReLU (Rectified Linear Unit)** helps the model learn complex patterns.

---

### 4. Pooling Layer

- **Downsamples the output** from the previous layer.
- Example: **Max Pooling** selects the maximum value in a region, reducing spatial dimensions.

---

### 5. Fully Connected Layer

- **Outputs the final class scores**.
- Typically results in a **11L** output, where **L** is the number of classes.

---

###  Note on CNN Performance

CNN performance is influenced by:
- Type and **depth of layers**.
- **Layer arrangement** and architecture.
- **Activation functions** used.
- **Quality of training data**.


## Building a Perceptron-based Linear Regressor

---

### Goal

- **Build a linear regression model** using a **perceptron** (a neural network approach).

---

### Tool

- **TensorFlow** for implementation.

---

### Context

- This exercise builds on the **linear regression** concept from earlier chapters.

````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Define the number of points to generate
num_points = 1200

# Generate the data based on the equation y = mx + c
data = []
m = 0.2
c = 0.5

for i in range(num_points):
    # Generate 'x'
    x = np.random.normal(0.0, 0.8)

    # Generate some noise
    noise = np.random.normal(0.0, 0.04)

    # Compute 'y'
    y = m * x + c + noise
    data.append([x, y])

# Separate x and y
x_data = np.array([d[0] for d in data], dtype=np.float32)
y_data = np.array([d[1] for d in data], dtype=np.float32)

# Generate weights and biases for the perceptron
W = tf.Variable(tf.random.uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))

# Define equation for 'y'
def model(x):
    return W * x + b

# Define how to compute the loss
def loss_fn(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred - y_true))

# Define the optimizer
optimizer = tf.optimizers.SGD(0.5)

# Training loop
num_iterations = 10
num_columns = 3  # Number of plots per row
num_rows = (num_iterations + num_columns - 1) // num_columns  # Calculate number of rows

# Create a figure for plotting
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes to easily iterate over

for step in range(num_iterations):
    with tf.GradientTape() as tape:
        # Compute the predicted y
        y_pred = model(x_data)

        # Calculate loss
        loss = loss_fn(y_pred, y_data)

    # Compute gradients and update weights and biases
    grads = tape.gradient(loss, [W, b])
    optimizer.apply_gradients(zip(grads, [W, b]))

    # Plot the progress on each subplot
    ax = axes[step]  # Get the corresponding axis
    ax.plot(x_data, y_data, 'ro')  # Plot the input data
    ax.plot(x_data, W.numpy() * x_data + b.numpy())  # Plot the predicted output line
    ax.set_xlabel('Dimension 0')
    ax.set_ylabel('Dimension 1')
    ax.set_title(f'Iteration {step + 1}')

    # Annotate numerical values (W, b, loss)
    ax.text(0.05, 0.95, f'W = {W.numpy()[0]:.3f}', transform=ax.transAxes, fontsize=10, verticalalignment='top')
    ax.text(0.05, 0.85, f'b = {b.numpy()[0]:.3f}', transform=ax.transAxes, fontsize=10, verticalalignment='top')
    ax.text(0.05, 0.75, f'Loss = {loss.numpy():.3f}', transform=ax.transAxes, fontsize=10, verticalalignment='top')

# Remove any unused subplots (if num_iterations is less than num_columns * num_rows)
for i in range(num_iterations, len(axes)):
    fig.delaxes(axes[i])

# Adjust layout for better display
plt.tight_layout()
plt.show()
````

## General Pattern

---

### W (Weight)

- The weight **W** is updated during each iteration as the model learns from the data.
- Initially, **W** starts with a random value and gradually adjusts to find the **optimal weight** for the best fit.

---

### b (Bias)

- The bias **b** also starts with a random value (typically near zero) and is updated to **minimize the loss function**.

---

### Loss

- The **loss** is the error between the predicted output (**y_pred**) and the actual data (**y_data**).
- The loss decreases as the model improves its predictions over time.

---

## Iteration Breakdown

- **W = -0.0614**, **b = 0.4881**: The model starts with random initial values for **W** and **b**.
- **Loss = 0.7111**: The loss is relatively high, indicating a poor fit to the data at the beginning of training.

### Iteration 2

- **W = 0.1199**, **b = 0.4956**: The weight **W** has increased in magnitude, and the bias **b** has slightly changed.
- **Loss = 0.0494**: The loss has decreased significantly, suggesting that the model is beginning to make better predictions.

### Iteration 3

- **W = 0.1762**, **b = 0.4980**: The model continues to adjust both **W** and **b** as it moves towards better parameters.
- **Loss = 0.0062**: The loss is much lower, indicating the model is fitting the data better.

### Iterations 4-5

- **W** and **b** continue to change slightly, but the adjustments are smaller now.
- The loss keeps decreasing from **0.0021** to **0.0017**, indicating the model is converging towards the optimal parameters.

### Iterations 6-7

- **W** and **b** are now very close to their final values. For example, in iteration 6, **W = 0.2008** and **b = 0.4991**.
- The loss stabilizes around **0.0016**, and the improvements in the loss are getting very small. The model is nearly converged.

### Iterations 8-10

- The updates in **W** and **b** are extremely small (from **W = 0.2015** to **0.2016**), and the loss remains nearly constant.
- The model has nearly reached its optimal parameters, as the loss is now very small (**0.0016**).

---

### Convergence

- The model starts with large changes in both **W** and **b**, and the loss is high.
- As training progresses, the updates become smaller, and the loss decreases, showing the model is converging towards the best fit.
- By the 10th iteration, the loss is very small, and **W** and **b** have stabilized.

### Accuracy

- A low loss value (**0.0016** at iteration 10) indicates that the model is now **accurately predicting** the output for the given input data.
- However, the model could potentially improve further with more iterations, but the changes are minimal at this point.


## Building an Image Classifier Using a Single Layer Neural Network

---

### Goal

- **Build an image classifier** using a **single layer neural network**.

---

### Dataset

- **MNIST** dataset, which consists of **handwritten digits**.

---

### Tool

- **TensorFlow** for building the model.

---

### Task

- **Classify digits correctly** from **MNIST images**.


````python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

# Set the input directory manually (since we don't use argparse here)
input_dir = './cifar10_data'

# Load the CIFAR-10 data using tensorflow.keras
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Preprocess the data: Flatten the images to 3072 (32x32x3)
x_train = x_train.reshape(-1, 3072).astype(np.float32) / 255
x_test = x_test.reshape(-1, 3072).astype(np.float32) / 255

# One-hot encode the labels
y_train_one_hot = tf.keras.utils.to_categorical(y_train, 10)
y_test_one_hot = tf.keras.utils.to_categorical(y_test, 10)

# Build the model using tf.keras
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(3072,)),  # Input layer with 3072 units (flattened image)
    tf.keras.layers.Dense(10, activation='softmax')  # Output layer with 10 units (one for each class)
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train_one_hot, epochs=10, batch_size=90, validation_data=(x_test, y_test_one_hot))

# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(x_test, y_test_one_hot, verbose=0)

# Make predictions on the test set
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate precision, recall, and F1 score
precision = precision_score(y_test, y_pred_classes, average='weighted')
recall = recall_score(y_test, y_pred_classes, average='weighted')
f1 = f1_score(y_test, y_pred_classes, average='weighted')

# Print final metrics
print(f'\nFinal Metrics:')
print(f'Accuracy: {test_accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')

# Plot training history
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Accuracy plot
axes[0, 0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[0, 0].set_title('Accuracy over epochs')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend()

# Loss plot
axes[0, 1].plot(history.history['loss'], label='Train Loss')
axes[0, 1].plot(history.history['val_loss'], label='Validation Loss')
axes[0, 1].set_title('Loss over epochs')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Loss')
axes[0, 1].legend()

# Precision plot
axes[1, 0].plot([precision] * len(history.history['accuracy']), label='Precision (constant)')
axes[1, 0].set_title('Precision')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Precision')

# F1 Score plot
axes[1, 1].plot([f1] * len(history.history['accuracy']), label='F1 Score (constant)')
axes[1, 1].set_title('F1 Score')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('F1 Score')

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.datasets import cifar10

# Load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize the pixel values to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Data Augmentation to improve model generalization
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
datagen.fit(x_train)

# Build a Convolutional Neural Network (CNN) model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),  # Dropout to reduce overfitting
    layers.Dense(10, activation='softmax')
])

# Compile the model with Adam optimizer
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model using the augmented data
history = model.fit(datagen.flow(x_train, y_train, batch_size=64),
                    epochs=10,
                    validation_data=(x_test, y_test),
                    verbose=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=2)

# Predict on the test set
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute additional metrics
precision = precision_score(y_test, y_pred_classes, average='weighted', zero_division=1)
recall = recall_score(y_test, y_pred_classes, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_pred_classes, average='weighted', zero_division=1)

# Print final metrics
print("\nFinal Metrics:")
print(f"Accuracy: {test_accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Plot training and validation accuracy/loss
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Accuracy plot
axes[0, 0].plot(history.history['accuracy'], label='Training Accuracy')
axes[0, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
axes[0, 0].set_title('Accuracy')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend()

# Loss plot
axes[0, 1].plot(history.history['loss'], label='Training Loss')
axes[0, 1].plot(history.history['val_loss'], label='Validation Loss')
axes[0, 1].set_title('Loss')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Loss')
axes[0, 1].legend()

# Precision plot
axes[0, 2].plot([precision]*len(history.history['accuracy']), label='Precision')
axes[0, 2].set_title('Precision')
axes[0, 2].set_xlabel('Epochs')
axes[0, 2].set_ylabel('Precision')

# Recall plot
axes[1, 0].plot([recall]*len(history.history['accuracy']), label='Recall')
axes[1, 0].set_title('Recall')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Recall')

# F1 Score plot
axes[1, 1].plot([f1]*len(history.history['accuracy']), label='F1 Score')
axes[1, 1].set_title('F1 Score')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('F1 Score')

# Empty subplot for layout
axes[1, 2].axis('off')

# Show the plots
plt.tight_layout()
plt.show()
````
